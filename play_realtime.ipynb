```python
import cv2
from ultralytics import YOLO

# load model
model = YOLO("best.pt")

# open video file
video_path = 0
cap = cv2.VideoCapture(video_path)
names = model.model.names

# loop through the video frames

while cap.isOpened():
    success, frame = cap.read()

    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    if success:
        results = model(frame)

        for result in results[0].boxes.cpu().numpy():
            x1, y1, x2, y2 = map(int, result.xyxy[0])
            cls = names[int(result.cls[0])]
            conf = result.conf[0].round(2)

            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 3)
            cv2.putText(frame, cls, (x1, y1), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(255, 0, 0))
            cv2.putText(frame, str(conf), (x2, y2), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.6, color=(255, 0, 0))

        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        cv2.imshow("YOLOv8", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    else:
        break

cap.release()
cv2.destroyAllWindows()
```
